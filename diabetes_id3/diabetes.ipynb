```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
mainData = pd.read_csv('diabetes.csv')
```

```python
#lets see first 5 row of data
mainData.head()
```

```python
#lets do train using our id3 library by sklearn

#first lets test with size 0.3
train_X, valid_X, train_y, valid_y = train_test_split(mainData[set(mainData.columns)-{'Outcome'}], mainData.Outcome, test_size=0.3, random_state=123)

model = DecisionTreeClassifier()
model = model.fit(train_X, train_y)
model.score(train_X, train_y)
```

```python
#valid score with test size 0.3
model.score(valid_X, valid_y)
```

```python
#looks bad, lets try increase test size

train_X, valid_X, train_y, valid_y = train_test_split(mainData[set(mainData.columns)-{'Outcome'}], mainData.Outcome, test_size=0.5, random_state=123)

model = DecisionTreeClassifier()
model = model.fit(train_X, train_y)
model.score(train_X, train_y)
```

```python
#valid score with test size 0.5
model.score(valid_X, valid_y)
```

```python
#hmmm,, maybe decreasing test_size
train_X, valid_X, train_y, valid_y = train_test_split(mainData[set(mainData.columns)-{'Outcome'}], mainData.Outcome, test_size=0.2, random_state=123)

model = DecisionTreeClassifier()
model = model.fit(train_X, train_y)
model.score(train_X, train_y)
```

```python
#valid score with test size 0.2
model.score(valid_X, valid_y)
```
